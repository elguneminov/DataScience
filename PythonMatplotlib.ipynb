```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,20)
x
```

```python
y = np.arange(5,100,5)
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("Horizontal Axis")
plt.ylabel("Vertical Axis")
plt.grid(True)
plt.show()
```

```python
plt.plot(x,y, color='r', linestyle="-.",linewidth=5)
plt.title("Line Plot")
plt.xlabel("Horizontal Axis")
plt.ylabel("Vertical Axis")
plt.grid(True)
plt.show()
```

```python
y2 = y/2

plt.plot(x,y, color='r', linestyle="-.",linewidth=5)
plt.plot(x,y2, color='g', linestyle=":",linewidth=3)

plt.title("Line Plot")
plt.xlabel("Horizontal Axis")
plt.ylabel("Vertical Axis")
plt.grid(True)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y, color='r', linestyle="-.",linewidth=5)

plt.title("Line Plot")
plt.xlabel("Horizontal Axis")
plt.ylabel("Vertical Axis")
plt.grid(True)


plt.subplot(2,1,2)
plt.plot(x,y2, color='g', linestyle=":",linewidth=3)

plt.title("Line Plot")
plt.xlabel("Horizontal Axis")
plt.ylabel("Vertical Axis")
plt.grid(True)
plt.show()
```

```python
employee = {"Vivian":5, "John":3, "Jane": 6}
```

```python
names = list(employee.keys())
```

```python
values = list(employee.values())
```

```python
plt.bar(names, values)
```

```python
plt.bar(names, values)
plt.title("Bar Plot")
plt.xlabel("Employee Names")
plt.ylabel("Work Days")
plt.grid(True)
plt.show()
```

```python
plt.barh(names, values, color='r')
plt.title("Bar Plot")
plt.xlabel("Employee Names")
plt.ylabel("Work Days")
plt.grid(True)
plt.show()
```

```python
countries = ["USA", "Germany", "France"]
population = [300,220,180]
```
